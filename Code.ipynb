```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img =cv2.imread('redSeaBream.png')

blur = cv2.GaussianBlur(img, (5,5), 0)

mera_format = cv2.cvtColor(blur, cv2.COLOR_BGR2HSV)

lower_orange = np.array([5, 45, 8])
upper_orange = np.array([50, 255, 255])
mask = cv2.inRange(mera_format, lower_orange, upper_orange)



# plt.imshow(mask, cmap='gray')

# Define the kernel
kernel = np.ones((5,5),np.uint8)

# Perform erosion
# image = cv2.erode(mask, kernel, iterations = 1)
# plt.imshow(image, cmap='gray')
# plt.show()

# Perform dilation twice
image = cv2.dilate(mask, kernel, iterations = 2)
# plt.imshow(image, cmap='gray')
# plt.show()

# Perform erosion again
image = cv2.erode(image, kernel, iterations = 3)

# Perform erosion again
# image = cv2.erode(image, kernel, iterations = 1)


# Contour detection
contours, _ = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filtering contours based on area
filtered_contours = [cnt for cnt in contours if cv2.contourArea(cnt) > 1000]

# Create a mask
mask = np.zeros_like(image)
cv2.drawContours(mask, filtered_contours, -1, 255, thickness=cv2.FILLED)

# Apply the mask to the original image
result = cv2.bitwise_and(image, mask)

# Show the image
plt.imshow(result, cmap='gray')
plt.show()
```
